```python
sns.distplot(df['tip'],kde=True,bins=40) #kde default is True.
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') #Jointplot has several kinds such as reg,resid,kde,hex,scatter
```
