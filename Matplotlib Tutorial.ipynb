# Matplotlib Tutorial 

## Import And Basic Examples

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Linear & Polynomial Graphs 

```python
#plt.plot(x axis value,y axis value)
x = np.linspace(-5000,5000,10000)
y=x
plt.plot(x, y, 'r') 
plt.xlabel('X Axis') #x axis name
plt.ylabel('Y Axis') #y axis name 
plt.title('Linear Graph') #Graph Title
plt.show()
```

```python
x = np.linspace(-5000,5000,10000)
y=x**2
plt.plot(x, y, 'blue') 
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Polynomial Graph')
plt.show()
```

### Trigonometrik Graphs

```python
x = np.linspace(-360,360,1000)
y=np.sin(np.radians(x))
plt.plot(x, y, 'brown') 
plt.xlabel('X Axis') #x axis name
plt.ylabel('Y Axis') #y axis name
plt.title('sinx') #Graph Title
x = np.linspace(-360,360,1000)
y = x-x
plt.plot(x,y,"r--")
x = np.linspace(0,0,1000)
y = x+np.linspace(-1,1,1000)
plt.plot(x,y,"r--")
plt.show()
```

```python
x = np.linspace(-360,360,1000)
y=np.cos(np.radians(x))
plt.plot(x, y, 'yellow') 
plt.xlabel('X Axis') #x axis name
plt.ylabel('Y Axis') #y axis name
plt.title('cosx') #Graph Title
x = np.linspace(-360,360,1000)
y = x-x
plt.plot(x,y,"r--")
x = np.linspace(0,0,1000)
y = x+np.linspace(-1,1,1000)
plt.plot(x,y,"r--")
plt.show()
```

### Creating Multiplots

```python
# plt.subplot(nrows, ncols, plot number)
x = np.linspace(0,10,10)
y=x**2
plt.subplot(1,2,1)
plt.plot(x, y, 'b--') 
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
# one row and two column graph so 1x2
```

```python
x = np.linspace(0,10,10)
y=x**2
plt.subplot(2,1,1)
plt.plot(x, y, 'y--') 
plt.subplot(2,1,2)
plt.plot(y, x, 'r*-')
# two row and one column graph so 2x1
```

## Object Oriented Matplotlib

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1, 0.8, 0.9]) # left, bottom, width, height (range is between 0 and 1)

x = np.linspace(0,10,100)
y=x**2

axes.plot(x, y, 'b')
axes.set_xlabel('X Axis') 
axes.set_ylabel('Y Axis')
axes.set_title('y = x^2')
```

```python
#Create a Nested Graphs
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes2 = fig.add_axes([0.23, 0.53, 0.4, 0.3]) 

# Main Axes
axes1.plot(x, y, 'b')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('y = x^2')

# Insert Figure Axes 2
axes2.plot(x,y, 'r--')
axes2.set_xlabel('X label 2')
axes2.set_ylabel('Y label 2')
axes2.set_title('y = x^2');
```

```python
# Create a default figure
fig, axes = plt.subplots()

axes.plot(x, y, 'y')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('y = x^2');
```

```python
#Define size of any figure
fig, axes = plt.subplots(figsize=(9,6))

axes.plot(x, y, 'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('y = x^2')
```

```python
#Save figure
fig.savefig("Polynomial.png")
```

### Legends

```python
#Create a Label 
fig = plt.figure(figsize=(6,6))

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x^2", color="#8B008B")
ax.plot(x, x**3, label="x^3", color="#FF8C00")
ax.legend(loc = "center") #best, upper right/left, lower left/right, center are these locations of graph 
```

### Line And Marker Style

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="green", lw=3, linestyle='-')
ax.plot(x, x+2, color="green", lw=3, ls='-.')
ax.plot(x, x+3, color="green", lw=3, ls=':')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue",marker='+')
ax.plot(x, x+2, color="blue",marker='o')
ax.plot(x, x+3, color="blue",marker='s')
```

```python
x = np.linspace(0,10,20)
n = np.arange(0,10)
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between")
```

## Axis Grid

```python
x = np.linspace(0,10,20)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=1)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=4)
axes[1].grid(color='orange', alpha=0.5, linestyle='dashed', linewidth=0.8)
```

## Twin Axes

```python
fig, axes1 = plt.subplots()

axes1.plot(x, x**2, lw=1, color="blue",alpha=0.8)
axes1.set_ylabel(r"area $(m^2)$", fontsize=10, color="blue")
for label in axes1.get_yticklabels():
    label.set_color("blue")
    
axes2 = axes1.twinx()
axes2.plot(x, x**3, lw=1.5, color="red")
axes2.set_ylabel(r"volume $(m^3)$", fontsize=10, color="red")
for label in axes2.get_yticklabels():
    label.set_color("red")
```

## Coordinate Plane

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))  

x = np.linspace(-360, 360, 1000)
y = np.sin(np.radians(x))
ax.plot(x, y);
```
